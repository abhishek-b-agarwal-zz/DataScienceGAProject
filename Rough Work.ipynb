```python
import networkx as nx
graph = nx.Graph()
nodes_already_added = set()
for index, row in df_train.iterrows():
    user = row['user_id']
    item = row['item_id']
    if user not in nodes_already_added:
        graph.add_node(user)
        nodes_already_added.add(user)
    if item not in nodes_already_added:
        graph.add_node(item)
        nodes_already_added.add(item)
    graph.add_edge(user, item)
```

```python
import math
def amazon_weight(item):
    return math.sqrt(nx.degree(graph,item))

def amazon_cosine(item1,item2):
    return (len(list(nx.common_neighbors(graph,item1,item2)))) / amazon_weight(item1) / amazon_weight(item2)
```

```python
# to pass one item and get n no. of most similar items
def most_similar_items_amazon(item,cnt_items):
    scores = [(amazon_cosine(item,other_item),other_item) 
              for other_item in df_train.item_id.unique() if  item != other_item ]
    scores.sort()
    scores.reverse()
    return scores[0:cnt_items]
```

```python
# to pass one item and get n no. of most similar items
def most_similar_items_jaccard(item,cnt_items):
    similar_items = pd.DataFrame()
    for other_item in df_train.item_id.unique():
        if  item != other_item:
            jaccard_coef = nx.jaccard_coefficient(graph,[(item,other_item)])
            for u,v,p in jaccard_coef:
                similar_items = similar_items.append({
                        'Jaccard Coefficient':p,
                        'item1':u,
                        'item2':v
                    },ignore_index=True)
    similar_items.sort_values('Jaccard Coefficient',ascending=False,inplace='True')
    return similar_items[0:cnt_items]
```

```python
from math import sqrt
def pearson_correlation(person1,person2):
    #To get both rated items
    both_rated = {}
    for item in df_train[df_train.user_id == person1].item_id.unique():
        if item in df_train[df_train.user_id == person2].item_id.unique():
            both_rated[item] = 1
    
    #print both_rated
    
    number_of_ratings = len(both_rated)
    # Checking for number of ratings in common
    if number_of_ratings == 0:
        return 0
    
    if number_of_ratings > 2:
        print person1, person2, number_of_ratings
 
    # Add up all the preferences of each user
    person1_preferences_sum = sum([df_train[(df_train.item_id == item) & (df_train.user_id == person1)].rating.values for item in both_rated])
    
    person2_preferences_sum = sum([df_train[(df_train.item_id == item) & (df_train.user_id == person2)].rating.values for item in both_rated])
    
    # Sum up the squares of preferences of each user
    person1_square_preferences_sum = sum([pow(df_train[(df_train.item_id == item) & 
                                                          (df_train.user_id == person1)].rating.values,2) for item in both_rated])
    
    person2_square_preferences_sum = sum([pow(df_train[(df_train.item_id == item) & 
                                                          (df_train.user_id == person2)].rating.values,2) for item in both_rated])
    
    # Sum up the product value of both preferences for each item
    product_sum_of_both_users = sum([df_train[(df_train.item_id == item) & (df_train.user_id == person1)].rating.values
                                     * df_train[(df_train.item_id == item) & (df_train.user_id == person2)].rating.values for item in both_rated])
    # Calculate the pearson score
    numerator_value = product_sum_of_both_users - (person1_preferences_sum*person2_preferences_sum/number_of_ratings)
    denominator_value = sqrt((person1_square_preferences_sum - 
                              pow(person1_preferences_sum,2)/number_of_ratings) * 
                             (person2_square_preferences_sum -pow(person2_preferences_sum,2)/number_of_ratings))
    if denominator_value == 0:
        return 0
    else:
        r = numerator_value/denominator_value
        return r
```

```python
def most_similar_users(user,cnt_users):
    scores = [(pearson_correlation(user,other_user),other_user) 
              for other_user in df_train.user_id.unique() if  user != other_user ]
    scores.sort()
    scores.reverse()
    return scores[0:cnt_users]
```
