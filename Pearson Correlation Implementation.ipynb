```python
from math import sqrt
def similarity_score(person1,person2):

# Returns ratio Euclidean distance score of person1 and person2 
 
    both_viewed = {}# To get both rated items by person1 and person2
 
    for item in rating_final[rating_final.user_id == person1].item_id.unique():
        if item in rating_final[rating_final.user_id == person2].item_id.unique():
            both_viewed[item] = 1
    
    print both_viewed

# Conditions to check they both have an common rating items	
    if len(both_viewed) == 0:
        return 0

# Finding Euclidean distance 
    sum_of_eclidean_distance = []

    for item in rating_final[rating_final.user_id == person1].item_id.unique():
        if item in rating_final[rating_final.user_id == person2].item_id.unique():
            sum_of_eclidean_distance.append(pow(rating_final[(rating_final.item_id == item) & (rating_final.user_id == person1)].rating.values 
                                                - rating_final[(rating_final.item_id == item) & (rating_final.user_id == person2)].rating.values,2))
            sum_of_eclidean_distance = sum(sum_of_eclidean_distance)
        return 1/(1+sqrt(float(sum_of_eclidean_distance)))
        #return sum_of_eclidean_distance
        

def pearson_correlation(person1,person2):
    #To get both rated items
    both_rated = {}
    for item in rating_final[rating_final.user_id == person1].item_id.unique():
        if item in rating_final[rating_final.user_id == person2].item_id.unique():
            both_rated[item] = 1
    
    #print both_rated
    
    number_of_ratings = len(both_rated)
    # Checking for number of ratings in common
    if number_of_ratings == 0:
        return 0
    
    if number_of_ratings > 2:
        print person1, person2, number_of_ratings
 
    # Add up all the preferences of each user
    person1_preferences_sum = sum([rating_final[(rating_final.item_id == item) & (rating_final.user_id == person1)].rating.values for item in both_rated])
    
    person2_preferences_sum = sum([rating_final[(rating_final.item_id == item) & (rating_final.user_id == person2)].rating.values for item in both_rated])
    
    # Sum up the squares of preferences of each user
    person1_square_preferences_sum = sum([pow(rating_final[(rating_final.item_id == item) & 
                                                          (rating_final.user_id == person1)].rating.values,2) for item in both_rated])
    
    person2_square_preferences_sum = sum([pow(rating_final[(rating_final.item_id == item) & 
                                                          (rating_final.user_id == person2)].rating.values,2) for item in both_rated])
    
    # Sum up the product value of both preferences for each item
    product_sum_of_both_users = sum([rating_final[(rating_final.item_id == item) & (rating_final.user_id == person1)].rating.values
                                     * rating_final[(rating_final.item_id == item) & (rating_final.user_id == person2)].rating.values for item in both_rated])
    # Calculate the pearson score
    numerator_value = product_sum_of_both_users - (person1_preferences_sum*person2_preferences_sum/number_of_ratings)
    denominator_value = sqrt((person1_square_preferences_sum - 
                              pow(person1_preferences_sum,2)/number_of_ratings) * 
                             (person2_square_preferences_sum -pow(person2_preferences_sum,2)/number_of_ratings))
    if denominator_value == 0:
        return 0
    else:
        r = numerator_value/denominator_value
        return r
```

```python
def user_reommendations(person):

    # Gets recommendations for a person by using a weighted average of every other user's rankings
    totals = {}
    simSums = {}
    rankings_list =[]
    for other in rating_final.user_id.unique():
        # don't compare me to myself
        if other == person:
            continue
        sim = pearson_correlation(person,other)
        if sim > 0:
            print other, sim
            
    # ignore scores of zero or lower
        if sim <= 0: 
            continue
        for item in rating_final[rating_final.user_id == other].item_id.unique():
            # only score movies i haven't seen yet
            if item not in rating_final[rating_final.user_id == person].item_id.unique():

                # Similrity * score
                totals.setdefault(item,0)                
                totals[item] += rating_final[(rating_final.item_id == item) & (rating_final.user_id == other)].rating.values * sim
                # sum of similarities
                simSums.setdefault(item,0)
                simSums[item]+= sim

    # Create the normalized list

    rankings = [(total/simSums[item],item) for item,total in totals.items()]
    rankings.sort()
    rankings.reverse()
    # returns the recommended items
    recommendataions_list = [recommend_item for score,recommend_item in rankings[0:3]]
    return recommendataions_list
```
